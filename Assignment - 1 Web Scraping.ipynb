```python
soup = BeautifulSoup(page.content)
soup
```

```python
header = []

for title in soup.find_all('span', class_ = "mw-headline"):
    header.append(title.text)
    
header
```

```python
# Answer - 3

dineout = requests.get('https://www.dineout.co.in/delhi-restaurants')
dineout
```

```python
soup = BeautifulSoup(dineout.content)
soup
```

```python
# (i)

rest_name = []

for name in soup.find_all('a', class_ = "restnt-name ellipsis"):
    rest_name.append(name.text)
    
rest_name
```

```python
# (ii)

cuisine = []

for types in soup.find_all('div', class_ = "detail-info"):
    cuisine.append(types.text)
    
cuisine
```

```python
# (iii)

location  = []

for place in soup.find_all('div', class_ = "restnt-loc ellipsis"):
    location.append(place.text)
    
location
```

```python
# (v)

img_url = soup.find('div', class_ = 'img-wrap')
img_url
```

```python
# Answer - 4

page = requests.get('https://presidentofindia.nic.in/former-presidents')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
names = []

for presidents in soup.find_all('div', class_ = "desc-sec"):
    names.append(presidents.text)
    
names
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Former Presidents of India' : names})
df
```
