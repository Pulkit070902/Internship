```python
!pip install selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome()
```

```python
# Answer - 1

driver.get('https://www.naukri.com/')
```

```python
designation = driver.find_element(By.CLASS_NAME, "suggestor-input")
designation.send_keys('Data Scientist')
```

```python
search = driver.find_element(By.CLASS_NAME, "qsbSubmit")
search.click()
```

```python
job_title = []
job_loc = []
company = []
exp_req = []
```

```python
# Scraping Job Title

title_tags = driver.find_elements(By.XPATH, '//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in title_tags[:10]:
    name = i.text
    job_title.append(name)
    
    
# Scraping Job Location
    
loc_tags = driver.find_elements(By.XPATH, '//span[@class="locWdth"]')
for i in loc_tags[:10]:
    loc = i.text
    job_loc.append(loc)
    

# Scraping the Company Name

comp_tags = driver.find_elements(By.XPATH, '//div[@class=" row2"]/span/a[1]')
for i in comp_tags[:10]:
    comp = i.text
    company.append(comp)
    
    
# Scraping Experience Required

exp_tags = driver.find_elements(By.XPATH, '//span[@class="expwdth"]')
for i in exp_tags[:10]:
    exp = i.text
    exp_req.append(exp)
```

```python
print(len(job_title), len(job_loc), len(company), len(exp_req))
```

```python
df = pd.DataFrame({'Designation' : job_title, 
                   'Location' : job_loc, 
                   'Company Nmae' : company, 
                   'Experince' : exp_req})
df
```

```python
# Answer - 6

driver.get('https://www.azquotes.com/')
```

```python
quote = []
author = []
quote_type = []
```

```python
# Scraping Quote title

start_pg = 0
end_pg = 10

for page in range(start_pg, end_pg):
    title = driver.find_elements(By.XPATH, '//a[@class="title"]')
    for i in title:
        quote.append(i.text)
        
next_button = driver.find_element(By.XPATH, '//li[@class="next"]')
next_button.click()


# Scraping Quote Author

start_pg = 0
end_pg = 10

for page in range(start_pg, end_pg):
    name = driver.find_elements(By.XPATH, '//div[@class="author"]/a')
    for i in name:
        author.append(i.text)
        
next_button = driver.find_element(By.XPATH, '//li[@class="next"]')
next_button.click()


# Scraping Quote Type

start_pg = 0
end_pg = 10

for page in range(start_pg, end_pg):
    Q_type = driver.find_elements(By.XPATH, '//div[@class="tags"]')
    for i in Q_type:
        quote_type.append(i.text)
        
next_button = driver.find_element(By.XPATH, '//li[@class="next"]')
next_button.click()
```

```python
print('Title \n', quote, '\n', '-'*250, '\n')

print('Author \n', author, '\n\n', '-'*250, '\n')

print('Quote type \n', quote_type, '\n\n', '-'*250, '\n')
```
